### Financial Features Analysis

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_excel('/Users/giovanni/Documents/QVF - quantitative framework for valuation multiples computation in mergers and acquisitions/data/00_raw/sic_industrial_and_commercial_machinery_and_computer_equipment.xlsx')

# Univariate Analysis
def univariate_analysis(data):
    # Summary statistics for numerical features
    numerical_summary = data.describe()
    print("Summary Statistics for Numerical Features:")
    print(numerical_summary)

    # Histograms for numerical features
    data.hist(figsize=(12, 8))
    plt.suptitle("Histograms of Numerical Features")
    plt.show()

    # Bar plot for categorical features
    categorical_features = data.select_dtypes(include=['object']).columns
    if len(categorical_features) > 0:
        for feature in categorical_features:
            sns.countplot(x=feature, data=data)
            plt.title(f"Count of {feature}")
            plt.show()

# Bivariate Analysis
def bivariate_analysis(data):
    # Pairplot for numerical features
    sns.pairplot(data)
    plt.suptitle("Pairplot of Numerical Features")
    plt.show()

# Correlation Analysis
def correlation_analysis(data, save_path=None):
    # Correlation matrix
    correlation_matrix = data.corr()

    # Heatmap of correlation matrix
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
    plt.title("Correlation Heatmap")
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path)
    plt.show()
    
univariate_analysis(data)

bivariate_analysis(data)

correlation_analysis(data, save_path="correlation_heatmap.png")
```

### F-statistic test

```python
from scipy.stats import f_oneway

def f_test_analysis(data):
    result = f_oneway(*[data[feature] for feature in data.columns])
    return result

f_test_results = f_test_analysis(data)

print("F-Test Results:")
print(f_test_results)
```
